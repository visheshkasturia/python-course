```python
class Node:
    def __init__(self,vertex):
        self.vertex=vertex
        self.rank=0
        self.parent=self
        

class Edge:
    def __init__(self,src,dest,wt):
        self.src=src
        self.dest=dest
        self.wt=wt

        
def findAndCompress(treeNode):
    if treeNode.parent==treeNode:
        return treeNode
    
    treeNode.parent=findAndCompress(treeNode.parent)
    return treeNode.parent



def union(treeNode1,treeNode2):
    if treeNode1.rank>treeNode2.rank:
        treeNode2.parent=treeNode1
    elif treeNode2.rank>treeNode1.rank:
        treeNode1.parent=treeNode2
    else:
        treeNode2.parent=treeNode1
        treeNode1.rank+=1
    
    return

def kruskal(edges,nVertices):
    edges=sorted(edges,key=lambda edge:edge.wt)
    count=0
    i=0
    output=[]
#     print(edges)
    while count<nVertices-1:
        currentEdge=edges[i]
        srcParent=findAndCompress(currentEdge.src)
        destParent=findAndCompress(currentEdge.dest)
        
#         print(srcParent.parent.vertex,end=" ")
#         print(destParent.parent.vertex)
        if srcParent!=destParent:
            count+=1
            output.append(currentEdge)
            union(srcParent,destParent)
#             print(srcParent.parent.vertex,end=" ")
#             print(destParent.parent.vertex)
            
            
        
        i+=1
    
    return output
    



# li=[int(x) for x in input().strip().split()]
# nVertices=li[0]
# nEdges=li[1]
# edges=[]
# vertices=[]
# for i in range(nVertices):
#     newNode=Node(i)
#     vertices.append(newNode)

# for i in range(nEdges):
#     s=[int(i) for i in input().strip().split()]
#     src=vertices[s[0]]
#     dest=vertices[s[1]]
#     wt=s[2]
#     newEdge=Edge(src,dest,wt)
#     edges.append(newEdge)

nVertices=6
nEdges=11
edges=[]
vertices=[]
for i in range(nVertices):
    newNode=Node(i)
    vertices.append(newNode)
newEdge1=Edge(vertices[0],vertices[2],5)
newEdge2=Edge(vertices[0],vertices[1],1)
newEdge3=Edge(vertices[1],vertices[3],3)
newEdge4=Edge(vertices[2],vertices[3],10)
newEdge5=Edge(vertices[2],vertices[1],6)
newEdge6=Edge(vertices[0],vertices[3],4)
newEdge7=Edge(vertices[2],vertices[4],8)
newEdge8=Edge(vertices[4],vertices[5],2)
newEdge9=Edge(vertices[5],vertices[3],6)
newEdge10=Edge(vertices[4],vertices[3],7)
newEdge11=Edge(vertices[2],vertices[5],9)
edges.append(newEdge1)
edges.append(newEdge2)
edges.append(newEdge3)
edges.append(newEdge4)
edges.append(newEdge5)
edges.append(newEdge6)
edges.append(newEdge7)
edges.append(newEdge8)
edges.append(newEdge9)
edges.append(newEdge10)
edges.append(newEdge11)

# print(edges)
# print(vertices)
output=kruskal(edges,nVertices)
for edge in output:
    print(str(edge.src.vertex)+ " "+ str(edge.dest.vertex) +" "+ str(edge.wt))
```
