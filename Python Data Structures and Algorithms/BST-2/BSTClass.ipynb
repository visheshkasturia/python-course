```python
import queue
class BinaryTreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None

class BST:
    
    def __init__(self):
        self.root=None
        self.numNodes=0
    
    def isPresentHelper(self,root,data):
        if root is None:
            return False
        
        if root.data==data:
            return True
        
        if data<root.data:
            return self.isPresentHelper(root.left,data)
        else:
            return self.isPresentHelper(root.right,data)
        
        
    def isPresent(self,data):
        return self.isPresentHelper(self.root,data)
    
    
    def insertHelper(self,root,data):
        
        if root is None:
            root=BinaryTreeNode(data)
            return root
        if data<root.data:
            root.left=self.insertHelper(root.left,data)
        else:
            root.right=self.insertHelper(root.right,data)
        
        return root
    
    def insert(self,data):
        self.root= self.insertHelper(self.root,data)
        self.numNodes+=1
    
    def findMinimum(self,root):
        if root is None:
            return 10000
        
        if root.left is None:
            return root.data
        else:
            return self.findMinimum(root.left)
    
    def deleteHelper(self,root,data):
        if root is None:
            return False,None
        
        if data<root.data:
            deleted, newLeft= self.deleteHelper(root.left,data)
            root.left=newLeft
            
            return deleted, root
        elif data>root.data:
            deleted, newRight=self.deleteHelper(root.right,data)
            root.right=newRight
            return deleted, root
        else:
            if root.right is None and root.left is None:
                return True,None
            elif (root.right is None and root.left is not None):
                return True, root.left
            elif (root.left is None and root.right is not None) :
                return True, root.right
            else:
                minimum=self.findMinimum(root.right)
                root.data=minimum
                deleted,newRightNode=self.deleteHelper(root.right,minimum)
                root.right=newRightNode
                
                return True, root
                
    def delete(self,data):
        deleted, newRoot= self.deleteHelper(self.root,data)
        self.root=newRoot
        if deleted:
            self.numNodes-=1
        return deleted
    
    def printHelper(self,root):
        q=queue.Queue()
        q.put(root)
        while (not(q.empty())):
            currentNode=q.get()
            print(currentNode.data,end=":")
            if currentNode.left!=None:
                print("L:"+str(currentNode.left.data), end=",")
                q.put(currentNode.left)
            else:
                print("L:-1", end=",")
            if currentNode.right!=None:
                print("R:"+str(currentNode.right.data), end="")
                q.put(currentNode.right)
            else:
                print("R:-1", end="")   
            print()
    
    def printBST(self):
        self.printHelper(self.root)
```

```python
b = BST()
for i in range(1,11,2):
    b.insert(i)

print(b.isPresent(8))
b.insert(-2)
b.insert(6)
b.delete(5)
b.delete(1)
b.printBST()
```
