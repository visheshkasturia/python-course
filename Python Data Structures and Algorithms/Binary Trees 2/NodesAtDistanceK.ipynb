```python
import queue
class BinaryTreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    

def nodesAtDepthK(root,k):
    if root is None:
        return
    if k==0:
        print(root.data)
        return
    
    nodesAtDepthK(root.left,k-1)
    nodesAtDepthK(root.right,k-1)


def printNodesAtDistanceK(root,k,element):
    
    if root is None:
        return -1
    
    if root.data==element:
        nodesAtDepthK(root,k)
        return 0
    
    
    leftDistance = printNodesAtDistanceK(root.left,k,element)
    if leftDistance != -1:
        leftDistance+=1
        if leftDistance==k:
            print(root.data)
        else:
            nodesAtDepthK(root.right,k-leftDistance-1)
        
        return leftDistance
    
    rightDistance=printNodesAtDistanceK(root.right,k,element)
    if rightDistance != -1:
        rightDistance+=1
        if rightDistance==k:
            print(root.data)
        else:
            nodesAtDepthK(root.left,k-rightDistance-1)

        return rightDistance
    
    return -1

def levelWiseTreeInput():
    li=[int(x) for x in input().strip().split()]
    if li[0]==1:
        return None
    rootData=li[0]
    root=BinaryTreeNode(rootData)
    q=queue.Queue()
    q.put(root)
    i=1
    while not(q.empty()):
        currentNode=q.get()
        if li[i]!=-1:
            leftChild=BinaryTreeNode(li[i])
            currentNode.left=leftChild
            q.put(leftChild)
        i+=1
        if li[i]!=-1:
            rightChild=BinaryTreeNode(li[i])
            currentNode.right=rightChild
            q.put(rightChild)
        i+=1
    
    return root
        

def printLevelWiseTree(root):
    q=queue.Queue()
    q.put(root)
    q.put(None)
    while not(q.empty()):
        currentNode=q.get()
        if currentNode!=None:
            print(currentNode.data,end=" ")
            if currentNode.left!=None:
                q.put(currentNode.left)
            if currentNode.right!=None:
                q.put(currentNode.right)
        else:
            if not(q.empty()):
                print()
                q.put(None)

root=levelWiseTreeInput()
printLevelWiseTree(root)
element=int(input())
k=int(input())
x=printNodesAtDistanceK(root,k,element)
```
