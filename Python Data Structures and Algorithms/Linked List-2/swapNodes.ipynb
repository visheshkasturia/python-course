```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def swap_nodes(head, i, j):
    swap1=head
    swap2=None
    curr=head
    swap1pre=None
    swap2pre=None
    while i>0 and j>0:
        swap1pre=curr
        curr=curr.next
        i-=1
        j-=1
    swap1=curr    
    while i>0:
        swap2pre=curr
        curr=curr.next
        swap2=curr
        i-=1
    while j>0:
        swap2pre=curr
        curr=curr.next
        swap2=curr
        j-=1
    
    
    if swap1.next==swap2:
        if swap1==head:
            head=swap2
        else:
            swap1pre.next=swap2
        swap1.next=swap2.next
        swap2.next=swap1    
    
    else:
        if swap1==head:
            head=swap2
        else:
            swap1pre.next=swap2

        temp=swap1.next
        swap2pre.next=swap1
        swap1.next=swap2.next
        swap2.next=temp
    
    return head
    
    
    

def ll(arr):
    if len(arr)==0:
        return None
    head = Node(arr[0])
    last = head
    for data in arr[1:]:
        last.next = Node(data)
        last = last.next
    return head

def printll(head):
    while head:
        print(head.data, end=' ')
        head = head.next
    print()

# Main
# Read the link list elements including -1
arr=list(int(i) for i in input().strip().split(' '))
# Create a Linked list after removing -1 from list
l = ll(arr[:-1])
i, j=list(int(i) for i in input().strip().split(' '))
l = swap_nodes(l, i, j)
printll(l)
```
