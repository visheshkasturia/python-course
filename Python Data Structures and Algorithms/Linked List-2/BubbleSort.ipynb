# Recursive

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

def takeInput():
    inputList=[int(x) for x in input().split()]
    
    head=None
    tail=None
    for currData in inputList:
        if currData==-1:
            break
        newNode=Node(currData)
        
        if head is None:
            head=newNode
            tail=newNode
        else:
            tail.next=newNode
            tail=newNode
    
    return head,tail

def printLL(head):
    while head is not None:
        print(str(head.data) + "->",end="")
        head=head.next
    print("None")

def swap(head):
    temp=head.next
    head.next=temp.next
    temp.next=head
    
    return temp
    
def bubbleSort(head,tail):
    
    curr=head
    prev=None
    ahead=head
    if head==tail:
        return head
    while curr.next != tail:
        ahead=curr.next
        if curr.data>ahead.data:
            if prev is None:
                head=swap(curr)
                curr=head.next
                prev=head
            else:
                prev.next=swap(curr)
                curr=prev.next.next
                prev=prev.next
        else:
            prev=curr
            curr=curr.next

    #Outside While
    ahead=curr.next
    if curr.data>ahead.data:
        if prev is None:
            head=swap(curr)
        else:
            prev.next=swap(curr)
    if prev is None:
        tail=head
    else:
        tail=prev.next
    head=bubbleSort(head,tail)
    return head
    
    
head,tail=takeInput()
head=bubbleSort(head,tail)
printLL(head)
```
